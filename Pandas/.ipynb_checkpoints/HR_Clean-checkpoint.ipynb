```python
df['IsActive'] = df['TerminationDate'].isna()
```

```python
df['TerminationDate'] = df['TerminationDate'].fillna('Still Employed')
```

```python
df.drop(columns=['TerminationDate'], inplace=True)
```

```python
df.drop(columns=['IsActive'], inplace=True)
```

```python
df["EngagementSurvey"] = df["EngagementSurvey"].astype(int)
df
```

```python
df.head(20)
```

```python
df['DateOfBirth'] = pd.to_datetime(df['DateOfBirth'])
df['HiringDate'] = pd.to_datetime(df['HiringDate'])
```

```python
df.isnull().sum()
```

```python
#Average salary by department:
df.groupby('Department')['Salary'].mean()
```

```python
#Active vs Terminated employee count:
df['EmploymentStatus'].value_counts()
```

```python
#Satisfaction vs Performance relationship:
df.groupby('PerformanceScore')['EmployeeSatisfaction'].mean()
```

```python
#Average tenure (years):
df['Tenure'] = (pd.Timestamp.today() - df['HiringDate']).dt.days / 365
df['Tenure'].mean()
```

```python
df.to_csv("empolyees.csv", index=False)
```
